```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'world-population.xlsm'
data = pd.read_excel(file_path)
data.head()

# Set the figure size for better visibility
plt.figure(figsize=(14, 7))

# Plotting the line chart
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(data['Year'], data['Population'], marker='o', linestyle='-', color='blue')
plt.title('World Population Over the Years (Line Chart)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(20, 8))
# Plotting the step chart
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.step(data['Year'], data['Population'], where='post', linestyle='-', color='purple')
plt.title('World Population Over the Years (Step Chart)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

plt.show()
plt.tight_layout()
```

```python
plt.figure(figsize=(20, 8))
# Plotting the line chart
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.plot(data['Year'], data['Population'], marker='*', linestyle='-', color='blue')
plt.title('World Population Over the Years (Line Chart)')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```
