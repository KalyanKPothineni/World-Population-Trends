```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate data for fill levels of bins over time
np.random.seed(42)  # For reproducibility
days = pd.date_range(start="2024-04-10", periods=7, freq='D')
bin_types = ['Organic', 'Recyclable', 'General']
data_fill_levels = {bin_type: np.random.randint(50, 100, size=len(days)) for bin_type in bin_types}
df_fill_levels = pd.DataFrame(data_fill_levels, index=days)

# Generate data for fuel consumption comparison
conditions = ['Before Optimization', 'After Optimization']
fuel_consumption = {'Fuel Consumption (Liters)': [np.random.randint(200, 300), np.random.randint(150, 250)]}
df_fuel_consumption = pd.DataFrame(fuel_consumption, index=conditions)

# Generate data for route efficiency improvement
route_efficiency = {'Distance Traveled (km)': [np.random.randint(500, 700), np.random.randint(400, 500)]}
df_route_efficiency = pd.DataFrame(route_efficiency, index=conditions)

# Plotting the graphs
plt.figure(figsize=(18, 5))

```

```python

# Fill levels over time
for bin_type in bin_types:
    plt.plot(df_fill_levels.index, df_fill_levels[bin_type], label=f'{bin_type} Bin')
plt.title('Fill Levels of Bins Over Time')
plt.xlabel('Date')
plt.ylabel('Fill Level (%)')
plt.legend()

```

```python

# Fuel consumption comparison
df_fuel_consumption['Fuel Consumption (Liters)'].plot(kind='bar', color=['blue', 'green'])
plt.title('Fuel Consumption Comparison')
plt.ylabel('Fuel Consumption (Liters)')
plt.show()
```

```python
# Route efficiency improvement
df_route_efficiency['Distance Traveled (km)'].plot(kind='bar', color=['blue', 'green'])
plt.title('Route Efficiency Improvement')
plt.ylabel('Distance Traveled (km)')
plt.show()
```
